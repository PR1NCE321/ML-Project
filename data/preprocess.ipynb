## Step 8
# Remove Irrelevant Columns

```python
data.drop(['LoanID'], axis=1, inplace=True)
```


## Step 9
# Identify Target Variable Distribution


```python
data['Default'].value_counts()
```


## Step 10
# Visualize Class Distribution


```python
sns.countplot(x='Default', data=data)
plt.show()
```


## Step 11
# Separate Numerical and Categorical Columns


```python
num_cols = data.select_dtypes(include=['int64','float64']).columns
cat_cols = data.select_dtypes(include=['object']).columns
```


## Step 12
# Outlier Detection Using Z-Score


```python
from scipy.stats import zscore
z_scores = np.abs(zscore(data[num_cols]))

Q1 = data[num_cols].quantile(0.25)
Q3 = data[num_cols].quantile(0.75)
IQR = Q3 - Q1
IQR
```

## Step 13
# Encoding Binary Categorical Variables

```python
binary_cols = ['HasMortgage','HasDependents','HasCoSigner']
for col in binary_cols:
 data[col] = data[col].map({'Yes':1,'No':0})
```

## Step 14
# Find numeric Column for Correlation

```python
num_df = data.select_dtypes(include=['int64', 'float64'])
```


## Step 15
# Compute Correlation Matrix

```python
corr_matrix = num_df.corr()
corr_matrix
```


## Step 16
# Visualize Correlation Matrix

```python

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## Step 17
# Simple Scatter Plot (Icome vs Loan Amount)

```python

# first 1000 record

plt.figure(figsize=(8,6))
plt.scatter(data['Income'][:1000], data['LoanAmount'][:1000])
plt.xlabel("Income")
plt.ylabel("Loan Amount")
plt.title("Income vs Loan Amount")
plt.show()
```

## Step 18 
# Line Graph



```python
df_sorted = data.sort_values(by='Income')

# first 500 record
plt.figure(figsize=(8,6))
plt.plot(df_sorted['Income'][:500], df_sorted['LoanAmount'][:500])
plt.xlabel("Income")
plt.ylabel("Loan Amount")
plt.title("Loan Amount Trend by Income")
plt.show()


























```

# Box Plot (Credit Score vs Loan Default)

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Default', y='CreditScore', data=data)
plt.xlabel("Loan Default")
plt.ylabel("Credit Score")
plt.title("Credit Score vs Loan Default")
plt.show()
```
